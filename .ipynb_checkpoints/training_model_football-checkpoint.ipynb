## Test with custom model

```python
# Run batched inference on a list of images
# Define path to video file
source = "image_test.jpg"

# Run inference on the source
results = model(source, stream=True)

# Process results generator
for result in results:
    boxes = result.boxes  # Boxes object for bounding box outputs
    masks = result.masks  # Masks object for segmentation masks outputs
    keypoints = result.keypoints  # Keypoints object for pose outputs
    probs = result.probs  # Probs object for classification outputs
    obb = result.obb  # Oriented boxes object for OBB outputs
    result.show()  # display to screen
    result.save(filename="result.jpg")  # save to disk
```

```python
# Define path to video file
source = "test_video.mp4"

# Run inference on the source
results = model(source, stream=True)  # generator of Results objects
```

```python
# Check if CUDA is available and choose the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load the model
model = YOLO("runs/detect/train/weights/best.pt")

# Move the model to the GPU (if available)
model.to(device)

# Define the video source
source = "test_video.mp4"

# Run inference on the source, using GPU if available
results = model(source, stream=True)  # This will run inference on GPU if model is on GPU

# Iterate over the results and display the frames in real-time
for result in results:
    frame = result.orig_img  # Get the processed frame with detections
    
    # Display the frame in a window (already in numpy format, no need to move to GPU)
    cv2.imshow("Processed Video", frame)
    
    # Wait for a key press to move to the next frame (with a small delay)
    key = cv2.waitKey(1) & 0xFF  # This allows the window to update
    if key == ord('q'):  # Press 'q' to exit the window
        break

# Close all OpenCV windows
cv2.destroyAllWindows()
```
